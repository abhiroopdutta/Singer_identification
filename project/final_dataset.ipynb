```python
singer_list = {}
with open("dataset.txt", "r") as filenames:
    for filename in filenames:
        singer_id = filename.split("_")[0]
        if singer_id in singer_list:
            singer_list[singer_id] += 1;
        else:
            singer_list[singer_id] = 1;

        
for singer, count in singer_list.items():
    if(count >= 6):
        with open("dataset.txt", "r") as filenames:
            c=0
            for filename in filenames:
                if(c==6):
                    c=0
                    break
                if(filename.split("_")[0] == singer):
                    with open("final_dataset.txt", "a") as file:
                        file.write(filename)
                        c+=1                 
```
